```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_ia0_ie01 = pd.read_csv("../data/Ia0_Ie01_Is0.txt",delim_whitespace=True,skiprows=2,names=["x","B"])
df_ia0_ie07 = pd.read_csv("../data/Ia0_Ie07_Is0.txt",delim_whitespace=True,skiprows=2,names=["x","B"])
df_ia6_ie01 = pd.read_csv("../data/Ia6_Ie01_Is0.txt",delim_whitespace=True,skiprows=2,names=["x","B"])
df_ia6_ie07 = pd.read_csv("../data/Ia6_Ie07_Is0.txt",delim_whitespace=True,skiprows=2,names=["x","B"])

plt.figure(dpi=200)
plt.plot(df_ia0_ie01["x"]/df_ia0_ie01["x"].max()*180,df_ia0_ie01["B"],label="Sem tensão de armadura")
plt.plot(df_ia6_ie01["x"]/df_ia6_ie01["x"].max()*180,df_ia6_ie01["B"],label="Com tensão de armadura")
plt.legend()
plt.ylabel("B [T]")
plt.xlabel("Posição angular [Graus]")
plt.title("Baixa corrente de excitação")
plt.grid()
plt.savefig("../img/baixa_iexc.png")

plt.figure(dpi=200)
plt.plot(df_ia0_ie07["x"]/df_ia0_ie07["x"].max()*180,df_ia0_ie07["B"],label="Sem tensão de armadura")
plt.plot(df_ia6_ie07["x"]/df_ia6_ie07["x"].max()*180,df_ia6_ie07["B"],label="Com tensão de armadura")
plt.ylabel("B [T]")
plt.xlabel("Posição angular [Graus]")
plt.title("Alta corrente de excitação")
plt.legend()
plt.grid()
plt.savefig("../img/alta_iexc.png")
```
